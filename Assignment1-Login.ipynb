```python
import re
import pandas as pd

def register():
    #Function to register a new user
    
    regex = '^[a-zA-Z]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,4}$' 

    fn=input("Enter your First Name :")
    ln=input("Enter your Last Name :")
    em=input("Enter email id (This will be the user name to login)  :")
    
    while True:
        if(re.search(regex,em)):   
            print("Valid Email")
            break
        else:   
            print("Invalid Email")
            em=input("Re-Enter email id :")

    pw=input("Enter password :")
    cp=input("Confirm password :")

    user=[]
    

    while True:
        
        reg = '^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*#?&])[A-Za-z\d@$!#%*?&]{5,15}$'
        
        if(pw == cp) and (re.search(reg,pw)):
            print("User Accepted!")
            user.append([fn, ln, em, pw])
            break    
        else:
            print("Invalid Password - Pls retry")
            pw=input("Enter password :")
            cp=input("Confirm password :")

    print(user)
    
    
    df=pd.DataFrame(user)
    df.to_csv(r'C:\Users\User\Downloads\Users.csv',mode='a',index=False,header=None)
    

    
def login():
    
    #Function to validate & login an existing user 
    
    uname=input("Enter username(email id) :")
    un=pd.read_csv(r'C:\Users\User\Downloads\Users.csv',usecols=[2])
    
    if uname in un.values:
        fgt=input("Do you remember your password (Y/N) :")
        if fgt=='Y':
            upw=input("Enter password :")
            pwd=pd.read_csv(r'C:\Users\User\Downloads\Users.csv',usecols=[3])
            if upw in pwd.values:
                print("Login successful")
            else:
                print("Login failed")
                print("Please register")
                register()
        else:
            df2=pd.read_csv(r'C:\Users\User\Downloads\Users.csv')
            p=df2.loc[df2['Email'] == uname, 'Password']
            print("Your password is:",p.to_string(index=False))
    else:
        print("Username not found")
        print("Please register")
        register()
    
chk=input("Are you a registered user (Y/N) :")
  


if chk=="N":
    register()
    
else:
    login()
    
```
